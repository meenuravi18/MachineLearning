```python
import pandas as pd
import numpy as np
# import webbrowser
import PyPDF2

import io

import requests
from PyPDF2 import PdfFileReader
```

```python
data=pd.read_csv("papers.csv")
data.head()
```

```python
missing_values=data.isna().sum()
missing_values
```

```python
data=data.drop(data.index[418])
```

```python
data=data.drop(['tl;dr'], axis=1)
data.head()
```

```python
missing_values=data.isna().sum()
missing_values
```

```python
data = data.replace(np.nan, '', regex=True)
```

```python
missing_values=data.isna().sum()
missing_values
```

```python
cleaned=pd.DataFrame()
```

```python
cleaned["len_title"]=data['title'].str.split().str.len()
cleaned["len_abstract"]=data['abstract'].str.split().str.len()
cleaned["num_authors"]=data['authors'].str.split(',').str.len()
cleaned["num_conflicts"]=data['conflicts'].str.split(',').str.len()
cleaned["num_tags"]=data['keywords'].str.split(',').str.len()
```

```python
cleaned.head()
```

```python
ivy=['brown', 'harvard', 'cornell', 'princeton', 'dartmouth', 'yale', 'columbia','upenn']
big_four=['facebook','amazon','google','apple']
ml_keywords=["natural language processing","unsupervised learning","supervised learning","neural network","deep learning",
            "reinforcement learning","computer vision","cluster","regression","classification"]
```

```python
cleaned['conflicts']=data['conflicts'].str.split(',')
```

```python
cleaned
```

```python
def ivy_conflict(s):
    affiliate=s.split(".",1)[0]
    if affiliate in ivy:
        return "yes"
    return "no"

def big_four_conflict(s):
    affiliate=s.split(".",1)[0]
    if affiliate in big_four:
        return "yes"
    return "no"
```

```python
cleaned['num_ivy_conflicts'] =  cleaned.conflicts.apply(lambda x: [ivy_conflict(i) for i in x].count("yes"))
cleaned['num_big_four_conflicts'] =  cleaned.conflicts.apply(lambda x: [big_four_conflict(i) for i in x].count("yes"))
```

```python
cleaned=cleaned.drop(['conflicts'], axis=1)
```

```python
cleaned
```

```python
abstract=data["abstract"].tolist()
num_keywords=[]
for i in abstract:
    count=0
    for j in ml_keywords:
        if j in i:
            count+=1
    num_keywords.append(count)
# num_keywords        
```

```python
cleaned["num_keywords_abstract"]=num_keywords
```

```python
title=data["title"].tolist()
num_keywords=[]
for i in title:
    count=0
    for j in ml_keywords:
        if j in i:
            count+=1
    num_keywords.append(count)
# num_keywords        
```

```python
cleaned["num_keywords_title"]=num_keywords
```

```python
cleaned
```

```python
cleaned["decision"]=data.decision.apply(lambda x: 'Accept' if 'Accept' in x else "Not Accept")
```

```python
cleaned
```

```python
pdfs=data["pdf_link"].tolist()
num_sections=[]
num_pages=[]
for i in pdfs:
    print(i)
    url = i
    if url !="https://openreview.net/pdf?id=HJWHIKqgl" and url!='https://www.dropbox.com/s/5ynhm1wy5vu2swq/DyVEDeep-ICLR2017.pdf':
        r = requests.get(url)
        f = io.BytesIO(r.content)
        reader = PdfFileReader(f)
        outlines = reader.getOutlines()
        sections=len(outlines)
        pages=reader.getNumPages()
        num_sections.append(sections)
        num_pages.append(pages)
    else:
        num_sections.append('NULL')
        num_pages.append('NULL')
```

```python
cleaned["num_pages"]=num_pages
cleaned["num_sections"]=num_sections
```

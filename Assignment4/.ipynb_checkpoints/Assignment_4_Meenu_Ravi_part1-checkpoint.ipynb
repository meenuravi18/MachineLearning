### Question 1: 15 points

#### - Load the MNIST dataset and split it into a training set and a test set: take the first 60,000 instances for training, and the remaining 10,000 for testing.



```python
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist["data"], mnist["target"]
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:] # the data is already shuffled
```

```python
print(X_train.shape)
print(y_test.shape)
```

#### - Train a Random Forest classifier on the dataset and time how long it takes, then evaluate the resulting model on the test set.


```python
t0 = time.time()
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
rnd_clf.fit(X_train, y_train)
t1 = time.time()

total = t1-t0
print(str(total)+" secs taken")
```

```python
y_pred_rf = rnd_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_rf)
```

#### - Next, use PCA to reduce the datasetâ€™s dimensionality, with an explained variance ratio of 95%.


```python
pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
d
```

```python
plt.figure(figsize=(6,4))
plt.plot(cumsum, linewidth=3)
plt.axis([0, 400, 0, 1])
plt.xlabel("Dimensions")
plt.ylabel("Explained Variance")
plt.plot([d, d], [0, 0.95], "k:")
plt.plot([0, d], [0.95, 0.95], "k:")
plt.plot(d, 0.95, "ko")
plt.annotate("Elbow", xy=(65, 0.85), xytext=(70, 0.7),
             arrowprops=dict(arrowstyle="->"), fontsize=16)
plt.grid(True)

plt.show()
```

```python
pca = PCA(n_components=154)
X_reduced = pca.fit_transform(X_train)
```

```python
X_reduced.shape
```

#### - Train a new Random Forest classifier on the reduced dataset and see how long it takes.


```python
t0 = time.time()
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
rnd_clf.fit(X_reduced, y_train)
t1 = time.time()

total = t1-t0
print(str(total)+" secs taken")
```

#### - Was training much faster?


The time taken on the training set when PCA was done was not faster. It took about 5 seconds with the original dataset and 12 seconds with the dimensions reduced. This shows that random forest classifier algorithm is not the most time efficient for this dataset and we should try a new classifier algorithm to achieve better time results

#### - Next, evaluate the classifier on the test set. How does it compare to the previous classifier?

```python
X_test_reduced = pca.transform(X_test)
y_pred_rf = rnd_clf.predict(X_test_reduced)
```

```python
accuracy_score(y_test, y_pred_rf)
```

Compared to the random forest classifier on the original dataset, the dimensionality reduced random forest classifier is less accurate. It looses around 5% accuracy. It is probably because when we reduce the dimensionality, we are considereing only dimensions that make up 95% of the variance, so some variance gets lost.

### Question 3: 15 points

#### The classic Olivetti faces dataset contains 400 grayscale 64Ã—64pixel images of faces. Each image is flattened to a 1D vector of size 4,096. 40 different people were photographed (10 times each), and the usual task is to train a model that can predict which person is represented in each picture.


#### Load the dataset using the sklearn.datasets.fetch_olivetti_faces() function, then split it into a training set, a validation set, and a test set (note that the dataset is already scaled between 0 and 1). Since the dataset is quite small, you probably want to use stratified sampling to ensure that there are the same number of images per person in each set.

```python
olivetti = fetch_olivetti_faces()
```

```python
X, y = olivetti["data"], olivetti["target"]
```

```python
print(X.shape)
print(y.shape)
```

```python
strat_split = StratifiedShuffleSplit(n_splits=1, test_size=40, random_state=42)
train_valid_idx, test_idx = next(strat_split.split(X, y))
X_train_valid = olivetti.data[train_valid_idx]
y_train_valid = olivetti.target[train_valid_idx]

X_test = olivetti.data[test_idx]
y_test = olivetti.target[test_idx]

strat_split = StratifiedShuffleSplit(n_splits=1, test_size=80, random_state=43)
train_idx, valid_idx = next(strat_split.split(X_train_valid, y_train_valid))
X_train = X_train_valid[train_idx]
y_train = y_train_valid[train_idx]
X_valid = X_train_valid[valid_idx]
y_valid = y_train_valid[valid_idx]
```

```python
X_train.shape
```

#### Next, cluster the images using K-Means, and ensure that you have a good number of clusters using silhouette score.

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X_train)
                for k in range(1, 280,10)]
```

```python
silhouette_scores = [silhouette_score(X_train, model.labels_)
                     for model in kmeans_per_k[1:]]
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(1, len(silhouette_scores)+1), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.show()
```

```python
maxi=0
ctr=1
bestk=0
for i in silhouette_scores:
    if i>maxi:
        maxi=i
    
for i in silhouette_scores:
    if i==maxi:
        bestk=ctr
    ctr+=1
print(bestk,maxi)   
```

```python
best_model=kmeans_per_k[bestk]
best_model
```

#### Visualize the clusters: do you see similar faces in each cluster?

```python
def plot_faces(faces, labels, n_cols=5):
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face.reshape(64, 64), cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()

for cluster_id in np.unique(best_model.labels_):
    print("Cluster", cluster_id)
    in_cluster = best_model.labels_==cluster_id
    faces = X_train[in_cluster].reshape(-1, 64, 64)
    labels = y_train[in_cluster]
    plot_faces(faces, labels)
```

I do see similar faces in each cluster. Especially the smiles, the glasses and nose shapes are very similar in each cluster

#### Search for the number of clusters that allows the classifier to get the best performance: what performance can you reach?

```python
clusters=[]
for n in range(1, 281,10):
    pipeline = Pipeline([
        ("kmeans", KMeans(n_clusters=n, random_state=42)),
        ("forest_clf", RandomForestClassifier(n_estimators=280, random_state=42))
    ])
    pipeline.fit(X_train_reduced, y_train)
    clusters.append((n, pipeline.score(X_valid_reduced, y_valid)))
    print(n, pipeline.score(X_valid_reduced, y_valid))
```

```python
maxi=0
ctr=0
max_ind=0
for i in clusters:
    if i[1]>maxi:
        maxi=i[1]
        max_ind=ctr
    ctr+=1
print(clusters[max_ind])
        
```
